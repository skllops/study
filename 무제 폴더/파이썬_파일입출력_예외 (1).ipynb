03 파일 읽기

작업디렉토리에 생성한 '매수종목1.txt' 파일을 읽은 후 종목코드를 리스트에 저장해보세요.

```
005930
005380
035420
```

````python
f = open("/Users/book/Desktop/뉴딜산업/매수종목1.txt", encoding="utf-8")
lines = f.readlines()   # python list           lines에 담아줌

codes = []                          #리스트생성
for line in lines:
    code = line.strip()               #'\n'  문자열 내부에 있는 공백 또는 특별한 문자를 구분해서, 리스트 아이템으로 만듦.
    codes.append(code)

print(codes)

f.close()
````

09 try, except, else, finally 구조 사용해보기

파이썬 예외처리는 다음과 같은 구조를 가질 수 있습니다.

```
try:
    실행 코드
except:
    예외가 발생했을 때 수행할 코드
else:
    예외가 발생하지 않았을 때 수행할 코드
finally:
    예외 발생 여부와 상관없이 항상 수행할 코드

```

아래의 코드에 대해서 예외처리를 사용하고 try, except, else, finally에 적당한 코드를 작성해봅시다. else와 finally는 적당한 문구를 print하시면 됩니다.

```
per = ["10.31", "", "8.00"]

for i in per:
    print(float(per))

```

````python
per = ["10.31", "", "8.00"]       #리스트

for i in per:
    try:
        print(float(i))         #i를 출력
    except:
        print(0)                #예외는 0
    else:
        print("clean data")     #아니면 클린데이터 출력
    finally:
        print("변환 완료")        #끝마치면 변환완료
````
