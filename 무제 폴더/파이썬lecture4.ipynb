```python
######################################2023.04.12#####################################
class car:
  def __init__(self,make,model,color,price):
    self.make = make                                    
    self.model = model
    self.color = color
    self.price = price
  def setmake(self,make):             #설정자 메소드
    self.make = make
  def getmake(self):                 #접근자 메소드
    return self.make

  def getdesc(self):                  #차량에 대한 정보를 문자열로 요약하여서 반환한다.
    return "차량" =("+str(self.make)+","+\
      str(self.model)+","+\
      str(self.color)+","+\
      str(self.price)+")"
```

```python
class electriccar(car):
  def __init__(self,make,model,color,price,batterysize):
    super().__init__(make,model,color,price)
    self.batterysize=batterysize
  def setbatterysize(self,batteryszie):
    self.batterysize = batterysize
  def getbatterysize(self):
    return self.batterysize

mycar = electriccar("tesla","model s","white",10000,0)
mycar.setmake("tesla")
mycar.setbatterysize(60)
print(mycar.getdesc())
```

```python
#########다중 상속############
class person:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def show(self):
    print(self.name,self.age)

class student:
  def __init__(self,id):
    self.id = id
  def getid(self):
    return self.id

class collegestudent(person,student):
  def __init__(self, name, age, id):
    person.__init__(self,name,age)
    student.__init__(self,id)

obj = collegestudent("kim",22,'1020202')
obj.show()
print(obj.getid())
```

```python
class employee:
  def __init__(self,name,salary):
    self.name = name
    self.salary = salary

  def getsalary(self):
    return self.salary

  def __str__(self):
    return f"이름: {self.name}, 월급: {self.getsalary()}"

class manager(employee):
  def __init__(self,name,salary,bonus):                 #여기에 밑에 값이 들어가
    super().__init__(name,salary)                       #name과 월급을 상속
    self.bonus = bonus
  def getsalary(self):
    return super().getsalary() + self.bonus             #위에 월급을 상속 + 셀프 보너스 옴?

kim = manager("김김",2000000000,2000000000)
print(kim)
```
